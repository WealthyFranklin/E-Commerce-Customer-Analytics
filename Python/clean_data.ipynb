```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('Online_Retail.csv')

# First look at your data
print(df.head())
print(df.info())
print(df.shape)
```

```python
# See which columns have missing values
print(df.isnull().sum())

# See percentage of missing values
print((df.isnull().sum() / len(df)) * 100)

# Focus on CustomerID and Description
print(f"Missing CustomerIDs: {df['CustomerID'].isnull().sum()}")
print(f"Missing Descriptions: {df['Description'].isnull().sum()}")
```

```python
# Remove rows where InvoiceNo starts with 'C'
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

print(f"Rows after removing cancellations: {len(df)}")
```

```python
# Remove negative quantities
df = df[df['Quantity'] > 0]

# Remove negative prices (assuming column is 'UnitPrice' or 'Price')
df = df[df['UnitPrice'] > 0]

print(f"Rows after removing negatives: {len(df)}")
```

```python
# Drop rows with missing CustomerID (can't analyze without customer info)
df = df.dropna(subset=['CustomerID'])

# Drop rows with missing Description (optional, depends on your analysis)
df = df.dropna(subset=['Description'])

print(f"Final row count: {len(df)}")
```

```python
# Reset the index after all the filtering
df = df.reset_index(drop=True)

# Save the cleaned data
df.to_csv('cleaned_data.csv', index=False)

# Final check
print("\nCleaned dataset info:")
print(df.info())
print(f"\nDate range: {df['InvoiceDate'].min()} to {df['InvoiceDate'].max()}")
print(f"Unique customers: {df['CustomerID'].nunique()}")
print(f"Unique products: {df['StockCode'].nunique()}")
```
